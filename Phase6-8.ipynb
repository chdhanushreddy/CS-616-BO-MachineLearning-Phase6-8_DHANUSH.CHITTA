```python
import keras
import tensorflow.keras as tk
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
from sklearn.metrics import roc_curve, auc
```

```python
# read data
(x_train, y_train), (x_test, y_test) = tk.datasets.mnist.load_data()
```

```python
# train data shape
print("train data shape")
print(x_train.shape)

print("test data shape")
print(x_test.shape)
```

```python
unique, counts = np.unique(y_train, return_counts=True)
plt.bar(unique, counts)
plt.title('Visualization of class distribution for the MNIST Dataset', fontsize=20, weight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Class', fontsize=16)
plt.ylabel('Frequency', fontsize=16);
```

```python
# train data
print("Train Images with number")
from matplotlib import pyplot as plt
fig, axs = plt.subplots(nrows=1, ncols=10, figsize=(10, 1))
for i, ax in enumerate(axs.flatten()):
    plt.sca(ax)
    plt.imshow(x_train[i], interpolation='nearest')
    plt.title(y_train[i])
plt.show()
```

```python
# preprocessing
x_train = x_train.reshape(x_train.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)
```

```python
# Create KNN model and fit model with training data
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x_train, y_train)
```

```python
y_pred = neigh.predict(x_test)
```

```python
# Accuracy
print("Accuracy")
print(accuracy_score(y_test, y_pred))
```

```python
# Confusion Matrix
print("Confusion Matrix")
print(confusion_matrix(y_pred, y_test))
```

```python
# Create KNN model and fit model with training data with optimization
neigh2 = KNeighborsClassifier(n_neighbors=5)
neigh2.fit(x_train, y_train)
y_pred2 = neigh2.predict(x_test)

# Accuracy
print("Accuracy")
print(accuracy_score(y_test, y_pred2))

# Confusion Matrix
print("Confusion Matrix")
print(confusion_matrix(y_pred2, y_test))
```

```python
# Create KNN model and fit model with training data with optimization
neigh3 = KNeighborsClassifier(n_neighbors=7)
neigh3.fit(x_train, y_train)
y_pred3 = neigh3.predict(x_test)

# Accuracy
print("Accuracy")
print(accuracy_score(y_test, y_pred3))

# Confusion Matrix
print("Confusion Matrix")
print(confusion_matrix(y_pred3, y_test))
```

```python
# Model Evaluation
train_x, test_x, train_y, test_y = train_test_split(x_train, y_train, test_size=0.33, random_state=42)
neigh4 = KNeighborsClassifier(n_neighbors=7)
neigh4.fit(train_x, train_y)
y_pred4 = neigh4.predict(test_x)

# Accuracy
print("Accuracy")
print(accuracy_score(test_y, y_pred4))

# Confusion Matrix
print("Confusion Matrix")
print(confusion_matrix(y_pred4, test_y))
```

```python
train_x, test_x, train_y, test_y = train_test_split(x_train, y_train, test_size=0.2, random_state=7)
neigh5 = KNeighborsClassifier(n_neighbors=3)
neigh5.fit(train_x, train_y)
y_pred5 = neigh5.predict(test_x)

# Accuracy
print("Accuracy")
print(accuracy_score(test_y, y_pred5))

# Confusion Matrix
print("Confusion Matrix")
print(confusion_matrix(y_pred5, test_y))
```
